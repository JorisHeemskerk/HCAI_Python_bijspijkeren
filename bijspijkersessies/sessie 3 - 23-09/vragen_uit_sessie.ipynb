```python
import random

# Generate a list of 700 random integers between 1 and 1000
random_numbers = [random.randint(1, 1000) for _ in range(700)]

print(random_numbers) # shape=700
```

```python
print(random_numbers[3: 5]) # index 5 niet geprint, want het is TOT, niet tot en met
```

```python
# eerste 7 getallen
print(random_numbers[0 : 7])
```

```python
correct_data = [] # 7, 100

for i in range(0, len(random_numbers), 7):
    correct_data.append(random_numbers[i : i+7])

print(correct_data)
```

```python
import numpy as np
```

```python
correct_data_np = np.array(correct_data)
print(correct_data_np)
```

```python
with open("dataset.csv", "w") as file:
    file.write("a,b,c,d,e,f,g\n")
    for row in correct_data:
        # print(row)
        for item in row:
            # print(item)
            file.write(str(item))
            file.write(",")
        file.write("\n")


print("done")
```
