```python
import pandas as pd
```

```python
df = pd.read_csv("category_splitsing.csv", sep=",")
df.head()
```

```python
def remove_university_suffix(category: str)-> str:
    university = category.replace('University', "").strip()
    return university
    
    
    # if "(B/T)" in category:
    #     return "B/T"
    # elif "(SOC)" in category:
    #     return "SOC"
    

remove_university_suffix("Harvard University")
```

```python
df["university"] = df["category"].apply(remove_university_suffix)
df.head()
```

```python
def average(two_columns)-> float:
    average = sum(two_columns) / 2
    return average
```

```python
df["average_count"] = df[["count_2024", "count_2025"]].apply(average, axis=1)
df.head()
```
