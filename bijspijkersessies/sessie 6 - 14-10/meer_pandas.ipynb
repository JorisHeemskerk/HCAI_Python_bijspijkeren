### Correlatie tussen age, height & weight

```python
import seaborn as sns
import numpy as np
```

```python
corr_matrix = df[["age", "height", "weight"]].corr()

# De matrix is van Pandas, die moet naar numpy
corr_matrix = corr_matrix.to_numpy()

np.fill_diagonal(corr_matrix, None)

sns.heatmap(corr_matrix, annot=True, cmap='coolwarm', xticklabels=["age", "height", "weight"], yticklabels=["age", "height", "weight"])
```
